# I. Planteamiento del problema
Las condiciones hipóxicas activan mecanismos moleculares y fisiológicos con el fin de inducir la termogénesis y mitigar los efectos negativos de la exposición prolongada a la falta de oxígeno (Kang et al., 2023). El ajuste del metabolismo para la optimización de oxígeno metabólico constituye un proceso clave durante estas adaptaciones. El factor inducible por hipoxia (HIF-1A) juega un rol central en la respuesta a estas condiciones, al inducir la expresión de genes involucrados en el metabolismo, la vascularización, el control del estrés oxidativo, la respuesta inflamatoria y la regulación del crecimiento celular (Baudel et al., 2017; M. Li et al., 2021a; Tang et al., 2023). Asimismo, son relevantes los genes que regulan la actividad de HIF-1A, como EGLN1, que lo marca para su degradación en condiciones de normoxia, o VEGF y UCP1, que favorecen su expresión y estabilización (An et al., 2024; Basse et al., 2017; Fusco et al., 2022).

Investigaciones a nivel de distintos tejidos de roedores expuestos a hipoxia han dado a conocer una expresión diferencial de múltiples genes en respuesta a esta condición, siendo HIF-1A y las vías que modula un blanco recurrente de regulación (Hao et al., 2017; Schmidt et al., 2017). En este contexto, el presente trabajo plantea la siguiente pregunta de investigación: ¿Los roedores expuestos a condiciones hipóxicas presentan patrones diferenciales de expresión génica semejantes a los observados en roedores expuestos a condiciones controladas de laboratorio?

# IV. Metodología

1. Obtención de datos

    Se trabajará en base a la data del estudio de Poetzsch (2025), cuyos datos de RNA-seq se encuentra depositados en el Archivo de Nucleótidos Europeo (ENA), bajo el código de accesión PRJEB64658.

2. Control de calidad de las lecturas

    Para evaluar la calidad inicial de las lecturas brutas, se usará FastQC 0.11.9 (Andrews, 2010), identificando problemas como extremos de baja calidad, contaminación por adaptadores y bases con puntajes Phred bajos. Luego, se empleará Trimmomatic 0.39 (Bolger et al., 2014) para filtrar y recortar lecturas, eliminando adaptadores y bases de baja calidad (Phred score < 20). Una alta calidad de lecturas, sin adaptadores ni bases con baja calidad, asegura la confiabilidad de los análisis posteriores. Si las lecturas presentan problemas significativos, las métricas de calidad permiten ajustar los parámetros de recorte.

3. Ensamblaje de novo del transcriptoma

    En ausencia de un genoma de referencia, se realizará un ensamblaje de novo usando Trinity 2.13.2 (Grabherr et al., 2011). También se llevará a cabo un ensamblaje guiado por referencia para comparación. Para esto se usarán los genomas de Sigmodon hispidus (GCA_041721565.1) y Phyllotis vaccarum (GCA_033318265.1). La calidad y completitud del ensamblaje se evaluarán con BUSCO 5.2.2 (Simão et al., 2015), comparando los transcriptos ensamblados con conjuntos de genes ortólogos conservados. Los resultados incluyen estadísticas como el número de transcriptos ensamblados, la longitud promedio, el N50 y el porcentaje de genes completos según BUSCO.

4. Anotación funcional de transcriptos

    Los transcriptos ensamblados serán anotados usando Trinotate (Bryant et al., 2017), que integra bases de datos como UniProt, Pfam y GO para asignar funciones putativas a los genes. Se identificará específicamente genes relacionados con la termorregulación y la hipoxia, basándose en estudios previos en modelos similares (Giorello et al., 2018; Storz et al., 2010; Valdez et al., 2015).
